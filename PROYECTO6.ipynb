# Importacion de datos y filtracion por los 5 primeros socios comerciales (exportaciones e importaciones)

Para el primer paso se van a importar las librerias que ya estamos acostumbrados a utlizar para el manejo de datos y los visualizadores, en este caso se importa pandas, numpy, matplotlib y seaborn

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

Se leen los datos, tanto las exportaciones de Colombia por los ultimos 20 años y con todos sus principales socios y tambien las importaciones de Colombia en los ultimos 20 años para despues realizar la visualizacion de los datos
Se guardaron en un raw de GitHub para poder importarlos y leerlos sin necesidad de tener el documento en algun computador.

```python
exp=pd.read_table('https://raw.githubusercontent.com/SantiagoVillalobosB/SantiagoVillalobosB.github.io/main/Trade_Map_-_List_of_importing_markets_for_a_product_exported_by_Colombia.txt')
imp=pd.read_table('https://raw.githubusercontent.com/SantiagoVillalobosB/SantiagoVillalobosB.github.io/main/Trade_Map_-_List_of_supplying_markets_for_a_product_imported_by_Colombia.txt')
```

Primera visualizacion de los datos como un DataFrame de la base de datos de exportaciones.
En este dataset, tanto el de exportaciones como el de importaciones las filas hacen referencia a cada uno de los paises con los que Colombia tienen comercio internacional, y las columnas hace referencia a el valor que se exporto y se importo respectivamente durante cada año, a partir del 2001 hasta el 2020.

```python
exp
```

```python
imp
```

Para este proyecto se hara todo el analisis y toda la visualizacion de datos con los 5 principales socios comerciales para el año 2020, se va a filtrar el dataset para quedarnos unicamente con los primeros 5 socios de Colombia tanto en el ambito de exportaciones como de importaciones

```python
exp_cinco = exp.sort_values('Exported value in 2020')
exp_cinco.tail(6)
```

```python
exp_cinco=exp.drop(exp[exp['Exported value in 2020']<1274000].index)
```

```python
exp_cinco.set_index('Importers', inplace=True)
```

En este caso ya tenemos los primeros 5 socios comerciales en el ambito de exportaciones para Colombia. Se ordeno primeramente de menor a mayor y tomamos los 5 mayores socios de Colombia, para despues crear un nuevo dataset unicamente con los socios principales.

Los primeros cinco paises socios comerciales son: 
1. Estados Unidos 
2. China 
3. Panama 
4. Ecuador 
5. Brasil

```python
imp_cinco=imp.sort_values('Imported value in 2020')
imp_cinco.tail(6)
```

```python
imp_cinco=imp.drop(imp[imp['Imported value in 2020']<1643891].index)
imp_cinco
```

Para el caso de las importaciones se realizo el mismo proceso que a las exportaciones pero para este caso obtenemos que los primeros cinco paises socios a los cuales Colombia importa son: 
1. Estados Unidos
2. China
3. Mexico 
4. Brasil 
5. Alemania
