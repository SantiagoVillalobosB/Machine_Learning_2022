# 3 parte outliers

Para revisar el tema de los outliers es esencial tener la base de datos unicamente con variables numericas debido a que es logico que unicamente se encuentren outliers en variables numericas, para este caso se realizara una limpieza y se utilizaran unicamente las variables que sean numericas y que tenga sentido logico utilizarlas.

````python
from scipy.spatial.distance import mahalanobis
from scipy.stats import chi2
data.info()
````

````python
data_outliers=data.drop(['name','host_name','neighbourhood_group','neighbourhood','latitude','longitude','room_type','last_review','reviews_per_month','id'], axis=1)
````

````python
mahal_distances = []

for row in data_outliers.to_numpy():
  mahal_distances.append(mahalanobis(row, data_outliers.mean(), np.linalg.inv(np.cov(data_outliers.values.T))))
k = data_outliers.shape[0]
data_outliers['mahal_distances'] = mahal_distances
data_outliers['p_value'] = 1 - chi2.cdf(data_outliers['mahal_distances'], k-1)
````

````python
data_outliers
````

````python
data_outliers[data_outliers['p_value']>0.01]
````

Despues de realizar el proceso de outliers se puede evidenciar que para ninguna variable numerica hay datos inconsistentes debido a que el p_valor con un nivel de confianza del 1% todos los datos lo cumplen es decir que no hay ningun dato que se considere outlier.

# 6 parte visualizacion de datos


Antes de grafica vamos a realizar un proceso para utilizar los limites de los rangos intercuartilicos, debido a que como hay tanta dispersion en los datos, unicamente nos vamos a quedar unicamente con el limite superior y mayor de los rangos intercuartilicos

````python
def IQR(df):
  iqr=df.quantile(q=0.75)-df.quantile(q=0.25)
  return [df.quantile(q=0.75)+1.5*iqr, df.quantile(q=0.25)-1.5*iqr]
data_variables_numericas=data[['price','minimum_nights','number_of_reviews','reviews_per_month' ]]
df_limites=pd.DataFrame()
for columna in data_variables_numericas.columns:
  df_limites[columna]= IQR (data_variables_numericas[columna])

df_limites['Limites']=['superior','inferior']
df_limites.set_index('Limites',inplace=True)
df_limites
````
